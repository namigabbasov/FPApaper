```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(8, 8))

# Outer Circle for AI
ai_circle = patches.Circle((0.5, 0.5), 0.4, color='lightblue', ec='black', label='AI (Artificial Intelligence)')
ax.add_patch(ai_circle)

# Middle Circle for ML
ml_circle = patches.Circle((0.5, 0.5), 0.3, color='lightgreen', ec='black', label='ML (Machine Learning)')
ax.add_patch(ml_circle)

# Inner Circle for Deep Learning
dl_circle = patches.Circle((0.5, 0.5), 0.2, color='lightcoral', ec='black', label='Deep Learning')
ax.add_patch(dl_circle)

# Setting aspect of the plot to be equal, so the circles aren't elliptical
ax.set_aspect('equal', 'box')

# Removing axis
ax.axis('off')

# Setting legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], loc='center')

plt.title("Hierarchical Relationship: AI, ML, and Deep Learning")
plt.show()

import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(8, 8))

# Outer Circle for AI
ai_circle = patches.Circle((0.5, 0.5), 0.4, color='lightblue', ec='black', label='AI (Artificial Intelligence)')
ax.add_patch(ai_circle)

# Middle Circle for ML
ml_circle = patches.Circle((0.5, 0.5), 0.3, color='lightgreen', ec='black', label='ML (Machine Learning)')
ax.add_patch(ml_circle)

# Inner Circle for Deep Learning
dl_circle = patches.Circle((0.5, 0.5), 0.2, color='lightcoral', ec='black', label='Deep Learning')
ax.add_patch(dl_circle)

# Annotations
ai_text = "Broad discipline of creating\nintelligent machines."
ml_text = "Machines learn from data\nwithout explicit programming."
dl_text = "Neural networks with\nmany layers."

ax.text(0.5, 0.68, ai_text, ha='center', fontsize=9, bbox=dict(boxstyle="round,pad=0.3", edgecolor="none", facecolor="aliceblue"))
ax.text(0.5, 0.5, ml_text, ha='center', fontsize=9, bbox=dict(boxstyle="round,pad=0.3", edgecolor="none", facecolor="honeydew"))
ax.text(0.5, 0.32, dl_text, ha='center', fontsize=9, bbox=dict(boxstyle="round,pad=0.3", edgecolor="none", facecolor="mistyrose"))

# Setting aspect of the plot to be equal, so the circles aren't elliptical
ax.set_aspect('equal', 'box')

# Removing axis
ax.axis('off')

# Setting legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], loc='upper left')

plt.title("Hierarchical Relationship: AI, ML, and Deep Learning")
plt.show()






```
